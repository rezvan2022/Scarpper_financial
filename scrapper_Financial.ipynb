```python
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.keys import Keys
import time
from selenium.webdriver.support.ui import Select
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.common.by import By
from selenium.webdriver.support import expected_conditions as EC



```

```python
brower = webdriver.Chrome(ChromeDriverManager().install())
brower.get('https://www.macrotrends.net/stocks/charts/JNJ/johnson-johnson/financial-ratios')
```

```python

table = brower.find_element_by_id('contentjqxgrid')
table1=table.get_attribute('innerText')
len(table1)
table2=list(table1)
table2=table1.splitlines()
type(table2)
len(table2)
table2

```

```python
col = brower.find_element_by_id('columntablejqxgrid')
col=col.get_attribute('innerText')
col=col.splitlines()
t=len(col)
t=int(t)


```

```python
import pandas as pd
df = pd.DataFrame (table2)
tt=len(df)
tt=tt/t
tt=int(tt)






```

```python

# importing library
import numpy as np
 
# initializing list
lst = df
 
# converting list to array
table3 = np.asarray(lst).reshape(tt,t)
 
# displaying list
#print ("List:", lst)
 
# displaying array
print ("arr: ", table3)
table3.shape
```

```python
import csv

with open('Jonson.csv','w') as f:
    writer = csv.writer(f)
    writer.writerows(table3)
```
